```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## <code>Create a black image to work</code>

```python
blackImg = np.zeros(shape = (512, 512, 3),
                    dtype = np.int16)
```

## <code>Get the Shape of the Image</code>

```python
blackImg.shape
```

## <code>Show it</code>

```python
plt.imshow(blackImg)
```

## <code>Draw a Cicle</code>
<p><code>The center is the first number on x-axis and second on y-axis</code></p>

```python
cv2.circle(img = blackImg,
           center = (400, 100),
           radius = 50,
           color = (0, 255, 0),
           thickness = 8)

plt.imshow(blackImg)
```

## <code>Filled Circle</code>

```python
cv2.circle(img = blackImg,
           center = (400, 200),
           radius = 50,
           color = (0, 255, 255),
           thickness = -1)

plt.imshow(blackImg)
```

## <code>Draw a Rectangle</code>

<code>The first number is on x-axis & the second on the y-axis

We need two points. One forup and diagonally down</code>

```python
cv2.rectangle(blackImg,
              pt1 = (200, 200),
              pt2 = (300, 300),
              color = (0, 0, 255),
              thickness = 5)

plt.imshow(blackImg)
```

## <code>Draw a Triangle</code>

```python
vertices = np.array([[10, 450],
                    [110, 350],
                    [180, 450]],
                    np.int32)

pts = vertices.reshape(-1, 1, 2)

cv2.polylines(blackImg,
              [pts],
              isClosed = True,
              color = (100, 100, 255),
              thickness = 3)

plt.imshow(blackImg)
```

## <code>Filled Rectangle</code>

```python
cv2.rectangle(blackImg,
              pt1 = (200, 25),
              pt2 = (300, 150),
              color = (0, 150, 105),
              thickness = -1)

plt.imshow(blackImg)
```

## <code>Filled Triangle</code>

```python
vertices = np.array([[10, 450],
                    [110, 350],
                    [180, 450]],
                   np.int32)

pts = vertices.reshape(-1, 1, 2)

cv2.fillPoly(blackImg,
             [pts],
             color = (100, 100, 255))

plt.imshow(blackImg)
```

## <code>Draw Line</code>

```python
cv2.line(blackImg,
         pt1 = (512, 0),
         pt2 = (0, 512),
         color = (100, 255, 100),
         thickness = 3)

plt.imshow(blackImg)
```

## <code>Write Text</code>

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(blackImg,
            text = 'MAR',
            org = (210, 500),
            fontFace = font,
            fontScale = 3,
            color = (100, 200, 250),
            thickness = 3,
            lineType = cv2.LINE_AA)

plt.figure(figsize = (14, 9))
plt.imshow(blackImg)
```
