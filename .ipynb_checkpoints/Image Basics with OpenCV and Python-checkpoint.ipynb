# <code>OpenCV</code>

## <code>Import Libraries</code>

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### <code>Get the with the imread<code>

### <code>A nebulosa Helix no infravermelho -</code> [ESO](https://www.eso.org/public/images/eso1205ec/) - [Nasa](https://apod.nasa.gov/apod/image/1609/Helix_SpitzerSchmidt_4655.jpg)

```python
img2 = cv2.imread('Helix_SpitzerSchmidt_960.jpg')
```

## <code>Always check the type<code>

```python
type(img2)
```

### <code>Get the image shape</code>

```python
img2.shape
```

### <code>Let's see the image with the imshow</code>

```python
plt.figure(figsize = (18, 5))

plt.imshow(img2)
```

#### <code>Until  now we working with Matplotlib and RGB.</code>

#### <code>OpenCV is reading the channel as RGB.<code>

#### <code><code>We will convert OpenCV to the channels of the photo.<code>

```python
imgFlix = cv2.cvtColor(img2,
                       cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (12, 5))

plt.imshow(imgFlix)
```

#### <code>Scale it to Gray and check the Shape

```python
img2Gray = cv2.imread('Helix_SpitzerSchmidt_960.jpg',
                      cv2.IMREAD_GRAYSCALE)
```

```python
img2Gray.shape
```

```python
plt.figure(figsize = (12, 5))

plt.imshow(img2Gray, cmap = 'gray')
```

#### <code>Resize the image</code>

<code>To resize the image we change the Height - Width --> Width - Height</code>

```python
imgNew = cv2.resize(imgFlix,
                    (800, 400))
```

```python
plt.figure(figsize = (10, 4))

plt.imshow(imgNew)
```

### <code>Resize with Ratio</code>

```python
widthRatio = 0.5
heightRatio = 0.5
```

```python
img3 = cv2.resize(imgFlix,
                  (0, 0),
                 imgFlix,
                 widthRatio,
                 heightRatio)
```

```python
plt.figure(figsize = (12, 5))

plt.imshow(img3)
```

```python
img3.shape
```

### <code>Flip on Horizontal Axis<code>

```python
img4 = cv2.flip(imgFlix, 0)

plt.figure(figsize = (12, 5))

plt.imshow(img4)
```

### <code>Flip on Vertical Axis<code>

```python
img4 = cv2.flip(imgFlix, 1)

plt.figure(figsize = (12, 5))

plt.imshow(img4)
```

### <code>Flip on Horizontal and on Vertical Axis<code>

```python
img4 = cv2.flip(imgFlix, -1)

plt.figure(figsize = (12, 5))

plt.imshow(img4)
```

### <code>Change the screen size</code>

```python
lastImg = plt.figure(figsize = (12, 7)) #Basta alterar os figsize
ax = lastImg.add_subplot(111)
ax.imshow(imgFlix)
```
